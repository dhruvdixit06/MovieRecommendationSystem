```python
import ast
def genre_convert(data):
    genre = []
    for i in ast.literal_eval(data):
        genre.append(i['name'])
    return genre    
```

```python
df['genres'] = df['genres'].apply(genre_convert)
```

```python
df['keywords'].iloc[0]
```

```python
import ast
def name_convert(data):
    keywords = []
    for i in ast.literal_eval(data):
        keywords.append(i['name'])
    return keywords 
```

```python
df['keywords'] = df['keywords'].apply(name_convert)
```

```python
df['cast'].iloc[0]
```

```python
def cast_names(data):
    actors = []
    counter = 0
    for i in ast.literal_eval(data):
        if counter != 3:
            actors.append(i['name'])
            counter += 1
    return actors        
```

```python
df['cast'] = df['cast'].apply(cast_names)
```

```python
df['crew'].iloc[0]
```

```python
def crew_names(data):
    director = []
    for i in ast.literal_eval(data):
        if i['job'] == "Director":
            director.append(i['name'])
            break
    return director        
    
```

```python
df['crew'] = df['crew'].apply(crew_names)
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```
