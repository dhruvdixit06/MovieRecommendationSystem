```python
import ast
def converting(data):
    genre = []
    for i in ast.literal_eval(data):
        genre.append(i['name'])
    return genre
```

```python
movies_data['genres'] = movies_data['genres'].apply(converting)
```

```python
movies_data['genres'] = movies_data['genres'].apply(lambda x:" ".join(x))
```

- We need a metric to score the movie.
- Calculate score of every movie.
- Sort the scores and recommend best rated movie to users
- We'll use IMDB weighted average score to score the movies

Weighted Rating (WR) =



- v is the number of votes garnered by the movie
- m is the minimum number of votes required for the movie to be in the chart (the prerequisite)
- R is the mean rating of the movie
- C is the mean rating of all the movies in the dataset

```python
#Calculate the number of votes garnered by the 80th percentile movie
m = movies_data['vote_count'].quantile(0.80)
m
```

```python
## Calculate C
C = movies_data['vote_average'].mean()
C
```

### NOW LET FILTER OUT THE MOVIE

```python
df = movies_data[movies_data['vote_count'] >= m]
```

```python
df.head()
```

```python
df.info()
```

```python
def weighted_rating(x,m=m,C=C):
    v = x['vote_count']
    R = x['vote_average']
    return ((v/v+m)*R) + ((m/v+m)*C)
```

```python
df['score'] = df.apply(weighted_rating,axis=1)
```

```python
df.head()
```

```python
df.drop(labels=['userId','timestamp','rating'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
new_df = df.drop_duplicates()
```

```python
new_df = new_df.sort_values('score', ascending=False)

new_df[["title", "vote_count", "vote_average", "score",'popularity']].head(30)
```

```python
# Plot top 10 movies
def plot():
    popularity = new_df.sort_values("score", ascending=False)
    plt.figure(figsize=(12, 6))
    plt.barh(popularity["title"].head(10), popularity["score"].head(10), align="center", color="skyblue")
    plt.gca().invert_yaxis()
    plt.title("Top 10 movies")
    plt.xlabel("Popularity")
    plt.show()
    

plot()
```
